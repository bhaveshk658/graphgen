```python
import graphgen
from graphgen.data import get_training_data
from graphgen.graph import Graph
from graphgen.graph import Node
from graphgen.generate.generate import to_merge, convert_to_graph
from graphgen.metric import evaluate_traces
import matplotlib.pyplot as plt
```

```python
# Get training data
path = "/Users/bkalisetti658/desktop/graphgen/data/InteractionDR1.0/recorded_trackfiles/DR_USA_Roundabout_EP"

box = [[960, 1150], [980, 1040]]
traces = get_training_data(1, path, box)
print(len(traces))
```

```python
# Plot data
print(len(traces))
for trace in traces:
    for point in trace:
        plt.scatter(point[0], point[1], c='b', alpha=0.2)
```

```python
# Clean data
trips = graphgen.data.data.clean(traces, 50, 1)
```

```python
# Plot cleaned data
for trip in trips:
    for point in trip:
        plt.scatter(point[0], point[1], c='b', alpha=0.2)
```

```python
# Convert each point to nodes
for trip in trips:
    for i in range(len(trip)):
        point = trip[i]
        node = Node(point[0], point[1], point[2])
        trip[i] = node
```

```python
G = convert_to_graph(trips)
```

```python
G.draw()
```
